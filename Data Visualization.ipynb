## Variables Types

Variables in statistics and data analysis can be classified into several types based on their characteristics and the kind of data they represent. Some common types of variables include:

### Categorical Variables: 
These variables represent categories or labels and have distinct groups or levels. They can be further classified into nominal and ordinal variables:

a) Nominal Variables: 
Categories with no inherent order, like colors or gender.

b) Ordinal Variables: Categories with a meaningful order, but the differences between categories may not be equal, like education levels (high school, college, etc.).

### Numerical Variables (Quantitative Variables):

These variables represent quantities and can be further divided into two types:

a) Continuous Variables: These variables can take any numerical value within a certain range. Examples include height, weight, and temperature.

b) Discrete Variables: These variables take on distinct, separate values. Examples include the number of siblings, number of cars, and number of pets.

### 1. Scatter plots

<img src='v1.png' style="border-radius: 10px"/>

A scatter plot, also known as a scatterplot or scatter diagram, is a graphical representation of individual data points in a two-dimensional coordinate system. It is commonly used to visualize the relationship between two variables and to identify patterns or trends within the data. Scatter plots are especially useful for exploring the nature of the relationship between numerical variables.

**Interpreting Scatter Plots:**
- The distribution of points on the scatter plot can reveal the nature of the relationship between the two variables.
- If the points generally form a pattern that slopes from the lower left to the upper right, it suggests a positive correlation: as one variable increases, the other tends to increase as well.
- If the points form a pattern that slopes from the upper left to the lower right, it indicates a negative correlation: as one variable increases, the other tends to decrease.
- If the points do not show a clear pattern or trend, the variables might be unrelated or exhibit a weak correlation.

**Scatter Plot Patterns:**
- Linear Relationship: When the points roughly form a straight line, it suggests a linear relationship between the two variables. This means that a change in one variable is associated with a consistent change in the other variable.
- Nonlinear Relationship: If the points form a curved pattern, it indicates a nonlinear relationship between the variables. This might be quadratic, exponential, logarithmic, or another nonlinear form.
- Clustered Points: If the points are clustered around certain regions, it might indicate the presence of groups or clusters within the data.
- Outliers: Scatter plots can help identify outliersâ€”data points that deviate significantly from the overall pattern. Outliers can affect the interpretation of the relationship between variables.


Scatter plots are a fundamental tool in exploratory data analysis and can provide valuable insights into the data's underlying relationships. They help researchers and analysts understand whether a correlation exists between variables, the strength of that correlation, and whether any outliers or patterns are present.

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
plt.title('Sepal Length vs. Sepal Width')
```

### Scatter plots with hue

In the context of scatter plots, "hue" refers to a visual encoding technique used to represent a categorical variable. While scatter plots are primarily used to display the relationship between two numerical variables, the introduction of hue allows you to incorporate a third categorical variable into the visualization.

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs. Sepal Width')
```

### Scatter plot with regression fit

A regression fit in a scatter plot refers to a statistical method that involves fitting a line or curve to the data points on the scatter plot. The goal of regression analysis is to model and quantify the relationship between two variables. The regression fit provides a way to understand the general trend or pattern in the data and make predictions based on the observed relationship.

```python
sns.regplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
plt.title('Sepal Length vs. Sepal Width with Regression Fit')
```

### 2. Line plots

A line plot, also known as a line chart or line graph, is a graphical representation that displays data points connected by straight lines. Line plots are particularly useful for showing the trend and changes in data over a continuous interval, such as time. They are widely used to visualize relationships, patterns, and fluctuations in numerical data. One common use of line plots is to analyze time series data, such as stock prices, temperature changes, or population growth, to understand historical trends and make predictions.

In summary, line plots are a powerful tool for visualizing and understanding the trends and patterns in numerical data. They are particularly effective for representing data that changes continuously over time or another continuous interval. Line plots offer insights into how variables evolve and interact, making them a fundamental tool in data analysis, statistics, and scientific research.

```python
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
plt.title('Sepal Length vs. Sepal Width')
```

```python
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs. Sepal Width')
```

### Bar Plots

A bar plot, also known as a bar chart or bar graph, is a graphical representation that uses rectangular bars to show data values for different categories or groups. Bar plots are widely used to compare and visualize categorical data, making them effective for showing the distribution, frequency, or relationship between different groups.

In a bar plot, the x-axis typically represents the categories or groups being compared, while the y-axis represents the values associated with each category. The length or height of each bar corresponds to the value being represented.


Bar plots are particularly effective for categorical data analysis and for presenting data to a broad audience. They are commonly used in fields such as statistics, business, social sciences, and more, to convey insights about data distributions and relationships.

```python
sns.barplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Sepal Length vs. Sepal Width')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from Seaborn
df = pd.read_csv("iris.csv")

# Create a FacetGrid for subplots
g = sns.FacetGrid(df, col="Species")

# Map a line plot to the FacetGrid
g.map(sns.lineplot, "SepalLengthCm", "SepalWidthCm", marker="o")

# Add titles to the subplots
g.set_titles(col_template="{col_name} Species")

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```
