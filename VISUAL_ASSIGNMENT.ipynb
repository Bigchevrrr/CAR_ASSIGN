```python
#import dependencies


import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cloths - cloths.csv')
```

```python
df
```

# PRODUCT CATEGORY VERSUS COST PER UNIT

```python
sns.barplot(y='Product Category', x='Cost Per Unit', data = df)
plt.title('A Product Category and Cost Per Unit')
```

# QUANTITY VERSUS COST PER UNIT

```python
sns.lineplot(x='Cost Per Unit', y='Quantity', data = df)
plt.title('Quantity and Cost Per Unit')
```

# SELLPER UNIT VERSUS COST PER UNIT

```python
sns.scatterplot(x='Cost Per Unit', y='Selling Per Unit', data = df)
plt.title('A plot of Selling Per Unit and Cost Per Unit')
```
