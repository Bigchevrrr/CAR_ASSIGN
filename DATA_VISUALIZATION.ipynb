```python
#import dependencies


import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('Iris - Iris.csv')
```

```python
iris
```

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data = iris)
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

```python
#HUE
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue = 'Species', data = iris)
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

```python
#SCATTER PLOT WITH REGRESSION FIT

sns.regplot(x='SepalLengthCm', y='SepalWidthCm', data = iris)
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

```python
#CHANGE MAKERS; USE ANOTHER MARKER INSTEAD OF ZERO


sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data = iris)
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

# SCATTER PLOT

```python
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm', data = iris)
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

```python
#change colour

sns.lineplot(x='SepalLengthCm', y='SepalWidthCm', hue = 'Species', data = iris)
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

# BAR PLOT

```python
sns.barplot(x='Species', y='SepalWidthCm', data = iris)
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

# CHANGING THE ORIENTATION OF THE BARPLOT

```python
sns.barplot(y='Species', x='SepalWidthCm', data = iris, orient = 'h')
plt.title('A plot of SepalLengthCm and SepalWidthCm')
```

# SUBPLOT

```python
#CREATE A FACETGRID FOR SUBPLOTS
g = sns.FacetGrid(iris, col = "Species")

#Map a lineplot to facetgrid
g.map(sns.lineplot, "SepalWidthCm", "PetalLengthCm", marker = "o")

#ADD TITLES TO THE SUBLOTS
g.set_titles(col_template="{col_name} Species")

#ADJUST LAYOUTS
plt.tight_layout()

#SHOW THE SUBPLOTS
plt.show()

```
