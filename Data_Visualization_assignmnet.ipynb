### 1) What are categorical variables?

```python
categorical variable are variables that represents categories or labels which can be further sub-divided into ordinal and 
nominal. it is a wordly/verbal representation of variables.
```

### 2) What are numerical variables?

```python
Numerical variables which is also known as quantitative variables, it is a figurative representation of variable which can be 
further sub-divided into discret and continuous variables.
Discret variables are fixed variable e.g, number of cars WHILE 
Continuous variables are variables with no static value value. the value of continous variables changes over time e.g, height of
an human, body temperature and so on.
```

### 3) When do we use each of the following plots?
- a) Scatter plot
- b) Lineplot
- c) Bar plot
- d) Sub plot

```python
SCATTER PLOT: We use scatter plot when we have two numerical variables
LINEPLOT: We use lineplot to plot two numerical variables as well.
BAR PLOT: We use barplot to plot both numerical and categorical variables
SUB PLOT: We use subplot to create multiple plots on a single figure. Given the number of roles and columns.
we also use both categorical and numerical variables to create subplot.
```

### 4) Using the Addidas sales data, explain the correlation between price per unit and Operting profit using the appropriate chart.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Addidas_sales - Addidas_sales.csv')
```

```python
data
```

```python
data1 = data.head(1000)
```

```python
data['Total Sales'].dtype
```

```python
sns.scatterplot(y='Price per Unit', x='Operating Profit', data = data1)
plt.title('Price per Unit and Operating Profit')
```

    The correlation between the price per unit and operating profit is that they are both increasing. So therefore, 
    the higher the price per unit, the higher the operating profit
